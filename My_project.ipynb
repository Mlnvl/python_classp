```
#print skewness of critic_score, total_sales, na_sales, jp_sales, pal_sales, other_sales
df[['critic_score', 'total_sales', 'na_sales', 'jp_sales', 'pal_sales', 'other_sales']].skew()
```

```
df['critic_score'] = df['critic_score'].fillna(df['critic_score'].median())
```

```
#calculate the total sales by the sum of regional sales
df['total_sales'] = df['total_sales'].fillna(df[['na_sales', 'pal_sales', 'other_sales']].sum(axis=1))
```

```
sum_sales = df[['na_sales', 'pal_sales', 'other_sales']].sum(axis=0)
sum_sales
```

```
sum_sales = df[['na_sales', 'pal_sales', 'other_sales']].sum(axis=0)
sum_sales
```

```
#change missing 'developer' with 'publisher'
df['developer'] = df['developer'].fillna(df['publisher'])
```

```
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```
df['critic_score'] = df['critic_score'].fillna(df['critic_score'].median())
```

```
genre_stats=df.groupby('genre').agg({
    'critic_score':'median',
    'total_sales':'sum'
})
sorted_genres_by_score=genre_stats.sort_values(by='critic_score',ascending=False)
sorted_genres_by_sales=genre_stats.sort_values(by='total_sales',ascending=False)
```

```
platforms = df.groupby(['title', 'console'])['total_sales'].sum().sort_values(ascending=False).head(10)
```

```
# Calculer la moyenne des ventes par genre
genre_means = df.groupby('genre')['na_sales'].transform('mean')

# Imputer les valeurs manquantes dans 'na_sales' en utilisant la moyenne du genre correspondant
df['na_sales'] = df['na_sales'].fillna(genre_means)
```

```
# Calculer la moyenne des ventes par genre
genre_means = df.groupby('genre')['jp_sales'].transform('mean')

# Imputer les valeurs manquantes dans 'na_sales' en utilisant la moyenne du genre correspondant
df['jp_sales'] = df['jp_sales'].fillna(genre_means)
```

```
# Calculer la moyenne des ventes par genre
genre_means = df.groupby('genre')['pal_sales'].transform('mean')

# Imputer les valeurs manquantes dans 'na_sales' en utilisant la moyenne du genre correspondant
df['pal_sales'] = df['pal_sales'].fillna(genre_means)
```

```
# Calculer la moyenne des ventes par genre
genre_means = df.groupby('genre')['other_sales'].transform('mean')

# Imputer les valeurs manquantes dans 'na_sales' en utilisant la moyenne du genre correspondant
df['other_sales'] = df['other_sales'].fillna(genre_means)
genre_means
```

```
df['critic_score'] = df['critic_score'].fillna(df['critic_score'].mean())
```
